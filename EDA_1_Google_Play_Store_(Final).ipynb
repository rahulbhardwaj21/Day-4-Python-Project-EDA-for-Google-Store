```python
#Create a function to replace M and K (Million and Thousand from the end) Final answer in KB but in numerical value
def size_process(item):
  if str(item)[-1]== 'M':
    res = float(str(item).replace('M', ''))
    res = res*1024
    return res
  elif str(item)[-1]== 'k':
    res = float(str(item).replace('k', ''))
    return res
  else:
    return str(np.NaN)
```

```python
#Apply Function
df_copy['Size'] = df_copy['Size'].apply(size_process)
```

**Question5**-Which category app users are reviewing the most?

```python
df10 = df_copy.groupby(['Category','App'])['Reviews'].sum().sort_values(ascending = False).reset_index()[:10]
df10
```

**Answer5**-**Facebook** of **Social Media**  category user are reviewing the most

**Insight**-Facebook of Social Media category user are reviewing the most
